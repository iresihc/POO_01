```python
class CuentaBancaria:
    def __init__(self, cuenta_no, fecha_de_apertura, balance, cliente):
        self.cuenta_no = cuenta_no
        self.fecha_de_apertura  = fecha_de_apertura
        self.balance = balance
        self.cliente = cliente

# atributos, métodos
    def deposito(self, monto):
        self.balance += monto ## un += es que no estas declarando la variable
        print(f"${monto} ha sido depositado a tu cuenta.") ##la f llama todo lo que pusiste antes, jala los datos y los imprime

    def seleccionar(self, monto):
        if monto > self.balance:
            print("Balance Insuficiente.")
        else:
            self.balance -= monto
            print(f"${monto} ha sido seleccionado de tu cuenta.")

    def verificar_balance(self):
        print(f"El balance actual es ${self.balance}.")

    def print_detalles(self):
        print("Nombre:", self.cliente)
        print("Numero cuenta:", self.cuenta_no)
        print("Fecha de apertura:", self.fecha_de_apertura)
        print(f"Balance: ${self.balance}\n") ## la n es para poder cerrarlo como string


cursor_1 = CuentaBancaria(2345, "01-01-2011", 1000, "Cesar")
cursor_2 = CuentaBancaria(1234, "11-03-2011", 2000, "Galindo")


print("Detalles:")
cursor_1.print_detalles() #este print es una definicion
cursor_2.print_detalles()


print("=============================") ##Esto significa que va a imprimir todas tus variables y no poner muchas veces el print
cursor_2.print_detalles()

cursor_2.deposito(1000)
cursor_2.verificar_balance()

cursor_2.seleccionar(5000)

cursor_2.seleccionar(3400)
cursor_2.verificar_balance()
```

```python
from datetime import datetime ##libreria que crea datos fechados

import yfinance as yf

class Asset:
    def __init__(self, ticker, start_date, end_date = datetime.today().strftime('%Y-%m-%d')):
      ##tenemos 3 argumentos porque self no cuenta como uno
      ##datatime.today es para que te de las fechas hasta el dia de hoy
      ##%y-%m-%d le estas indicando que solo quieres año, mes y dia

        self.ticker = ticker
        self.start_date = start_date
        self.end_date = end_date

    def get_data(self): ##nos baja un data frame
        df = yf.download(self.ticker, start = self.start_date, end = self.end_date) ##aqui estas bajando el ticker de yahoofinance
        return df

    def  simple_returns(self, adj = False, cum = False):
      ##adj son precios de cierre ajustados, pones false para que no lo asigne, solo para que lo tome como variable
      ##cum es suma acumulada y tampoco lo asignas, lo toma como variable
        df = self.get_data()
        if adj:
            simple_returns = df["Adj Close"].pct_change().dropna()
            ##toma del dataframetus precios ajuntados, con el punto ingresas a esa funcion, con pct_change calculas los rendimientos y con dropna eliminas el nan que es nada, o sea no asignados
        else:
            simple_returns = df["Close"].pct_change().dropna()
        if cum:
            simple_returns = (1 + simple_returns).cumprod() - 1
            ##con cumprod sacas el producto acumulado de todo esto
        return simple_returns
```
